```python
#Program 1: Use the merge sort algorithm to sort an array in order
def merge_sort(A):
    #Breaks apart the array into multiple pieaces, then take each compare and compare them
    if len(A) > 1:
        #Take mid of array and seperate them
        part = len(A)//2
        C = A[:part]
        D = A[part:]
        #Then analyze thos smaller pieces
        merge_sort(C)
        merge_sort(D)
        
        #I for left array, j for right array, and k for final array
        i,j,k = 0,0,0
        
        #Compares the values in array
        while i<len(C) and j<len(D):
            if C[i] < D[j]:
                A[k] = C[i]
                i+=1
            else:
                A[k] = D[j]
                j+=1
            k+=1
        
        #if C array still has elements after D reach end index, it will put the rest of the ordered values in array
        while i<len(C):
            A[k]=C[i]
            i+=1
            k+=1
        
        #if D array still has elements after C reach end index, it will put the rest of the ordered values in array
        while j<len(D):
            A[k]=D[j]
            j+=1
            k+=1
        return A
    else:
        return A
```

```python
#Program 2: Find 2nd largest element in an array
def second_largest(A):
    #Set the largest and 2nd largest value as the first value in array
    largest = secondLargest = A[0]
    #As I iterate through the list
    for i in range(len(A)):
        #if the largest value changes, then the previous largest value is a prime candidate for 2nd largest
        if A[i] > largest:
            secondLargest = largest
            largest = A[i]
    return secondLargest
```

```python
#This actually my 2nd atttempt at this program as I used a recursion method at first, which came with issues when doing larger
#arrays as python has a limit to the amount of recursions in one cell
def find_peak2(A):
    #If array is size 1, then the peak is that element
    if len(A)==1:
        return A[0]
    start=0
    stop=len(A)-1
    peak=0
    #Analyzes both sides of array, moving towards a central peak
    while(start<stop):
        if A[start]>A[stop]:
            peak = A[start]
            stop-=1
        else:
            peak = A[stop]
            start+=1
    return peak
```

```python
#Program 4: See where A[i]=i in an array
def Ai_equals_i(A):
    #used the binary search method
    counter = 0
    start = 0
    stop = len(A)-1
    #Basically is A[i]<i at mid in an ordered array, then it can't exist in the first half of the array, and vice vera for the 
    #2nd half
    while start < stop:    
        mid = (start+stop)//2
        if A[mid]<mid:
            start = mid
        elif A[mid]>mid:
            stop = mid
        else:
            for i in range(start,stop):
                if(A[i]==i):
                    counter+=1
                    print("i="+str(i)+" A[i]="+str(A[i]))
            break;
    return "Total values where A[i]=i: "+str(counter)
```

```python
#Redid Karatsuba, but reduce calls to the algorithm and put a counter so it can count the number of time the program is called
def Karatsuba_Alg(x,y,counter):
    counter+=1
    #print(counter)
    if(len(str(x))==1 or len(str(y))==1):
        return x*y,counter
    else:
        
        n = max(len(str(x)),len(str(y)))
        if(n%2==1):
            n+=1
        a = x//10**(n//2)
        b = x%10**(n//2)
        c = y//10**(n//2)
        d = y%10**(n//2)
        #print("n = "+str(n))
        #print("a="+str(a)+" b="+str(b)+" c="+str(c)+" d="+str(d))
        ac,counter = Karatsuba_Alg(a,c,counter)
        bd,counter = Karatsuba_Alg(b,d,counter)
        #bc = Karatsuba_Alg(b,c)
        #ad = Karatsuba_Alg(a,d)
        #print("ac="+str(a)+"*"+str(c)+"="+str(ac))
        #print("bd="+str(b)+"*"+str(d)+"="+str(bd))
        
        #bc+ad = (bc+ad+ac+bd) - ac-bd = a(c+d)+b(c+d) -ac-bd = (a+b)(c+d) -ac-bd
        bc_plus_ad,counter = Karatsuba_Alg(a+b,c+d,counter)
        bc_plus_ad = bc_plus_ad - ac - bd
        #bc_plus_ad = bc+ad
        #print("bc+ad=(a+b)(c+d)-ac-bd= ("+str(a)+"+"+str(b)+")("+str(c)+"+"+str(d)+")-"+str(ac)+"-"+str(bd)+"="+str(bc_plus_ad))
        result = (ac*10**(n)) + (bc_plus_ad*10**(n//2)) + bd
        #print("product="+str(ac)+"*"+str(10**n)+" + "+str(bc_plus_ad)+"*"+str(10**(n//2))+" + "+str(bd)+" = "+str(result))
        return result,counter
```
