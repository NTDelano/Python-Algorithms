```python
# Vanilla QuickSort-pivot is always first element

# partition
import random

def partition(A, left, r):
    pivot = A[left]  # the pivot value is the first
    print('pivot = '+str(pivot))
    i = left+1       
    #print('i='+str(i))

    for j in range(left+1, r):
        print('i='+str(i)+' j='+str(j))
        print('A[i]='+str(A[i])+' A[j]='+str(A[j]))
        if A[j] < pivot:
            print(str(A[j])+'<'+str(pivot))
            print(A)
            print('Switch A['+str(i)+']='+str(A[i])+' and A['+str(j)+']='+str(A[j]))
            #print('Switch '+str(A[i])+' and '+str(A[j]))
            A[j], A[i] = A[i], A[j]
            i = i+1
            print(A)
        else:
            print('do nothing and j++')
    A[left], A[i-1] = A[i-1], A[left]
    print(A)
    return i-1

def quicksort(A, left, r):
    print('quicksort')
    print(A)
    if left >= r:
        return

    i = left
    j = partition(A, left, r)   # Then do the partitioning
    print()
    print(i,j)

    quicksort(A, left, j)      # 2 recursive calls- this one from left to j-1
    quicksort(A, j+1, r)    # This one from j+1 to the end of the array                                             
    



A=[3,8,2,5,1,4,7,6]

quicksort(A,0,len(A))
print(A)
```
