```python
#imports
import random
from decimal import Decimal
import operator
import matplotlib.pyplot as plt
```

```python
#Program 1: Tuples
def tuples_fun(n):
    #First create two empty lists
    x=[]
    y=[]
    #n is the number of points
    for i in range(n):
        #Chooses some random value within [-1,1] for both x and y, then append to their respective list
        xval = random.uniform(-1,1)
        x.append(round(xval,2))
        yval = random.uniform(-1,1)
        y.append(round(yval,2))
    #Combine lists to make tuples
    a = list(zip(x, y))
    print(a)
    print('First tuple is '+str(a[0])+' and the last tuple is '+str(a[-1]))
    
    #Sort from least to greatest based on the x value
    Px = sorted(a, key=operator.itemgetter(0))
    print('Coordinates from index 2-8 after sorting all tuples by x-coordinate:')
    
    #Print index 2-8
    for i in range(2,9):
        print('Index i='+str(i)+' is '+str(Px[i]))
```

```python
#Test with 10 points
tuples_fun(10)
```

```python
#Program 2: Create a circle of points
def circle_of_points(n):
    #Expand the plot so it is easier to see
    plt.figure(figsize=(10, 10))
    x=[]
    y=[]
    #Go through process from before of creating tuples based on n points
    for i in range(n):
        xval = random.uniform(-1,1)
        x.append(round(xval,2))
        yval = random.uniform(-1,1)
        y.append(round(yval,2))
    a = list(zip(x, y))
    
    #Then for each point, or individual tuple, check that the distance of that point from (0,0) is less than or equal to 1
    for t in a:
        distance = ((t[0])**2 + (t[1])**2)**(1/2)
        if distance <= 1:
            plt.plot(t[0],t[1],'b.')
    plt.show()
```

```python
#Test with 100 points
circle_of_points(100)
```

```python
#Test with 1000 points
circle_of_points(1000)
```

```python
#Test with 10000 points
circle_of_points(10000)
```

```python
#program 3: Find the closest pairs of points using the brute force method, which will have a complexity of O(n^2)
def closest_pair_bf(A):
    #Initialize the first two points as being the closest pair
    MinDistance = ((A[0][0]-A[1][0])**2 + (A[0][1]-A[1][1])**2)**(1/2)
    P1,P2 = A[0],A[1]
    
    #check every pair of points starting from one index, and checking all the indexes after
    for i in range(len(A)-1):
        for j in range(i+1,len(A)):
            
            #Check distance between points
            distance = ((A[i][0]-A[j][0])**2 + (A[i][1]-A[j][1])**2)**(1/2)
            #if that point is smaller than the minimum, then those points will be recorded
            if distance < MinDistance:
                MinDistance = distance
                P1,P2 = A[i],A[j]
                
    return P1,P2,MinDistance
```

```python
tuple1 = [(0,2.1), (-1.3, 4.3), (5.7, 9), (-5, 4), (2.4, 6)]
closest_pair_bf(tuple1)
```
